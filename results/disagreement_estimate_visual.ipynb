```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("figure_data/disagreement_estimate.csv")
```

```python
# set font and theme
sns.reset_defaults()
sns.set_context("notebook", font_scale=1.6)



# add dashed black line from (0, 0) to (1, 1)
plt.plot([0, 1], [0, 1], linestyle="dashed", color="black", linewidth=2)

# plot values
ax = sns.lineplot(data=df, x="divisiveness_rate", y="estimated_divisiveness",
             hue="spamming_rate", palette=sns.color_palette("crest", as_cmap=True),
             linewidth=3, alpha=0.5)

# set axis labels
plt.xlabel("True Rate of Disagreement")
plt.ylabel("Estimated Proportion\nof Disagreements")

sns.despine()
ax.set_aspect(0.75)

# add legend
ax.legend(title="Spamming Rate", loc="center left", bbox_to_anchor=(1, 0.5))

# save to file
plt.savefig("disagreement_estimate.pdf", format="pdf", bbox_inches="tight")
plt.show()
```
