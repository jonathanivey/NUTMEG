```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
vis_df = pd.read_csv("figure_data/subpopulation_size.csv")

# only visualize NUTMEG results
vis_df = vis_df[vis_df['method'] == 'NUTMEG']

# only include odd numbered annotators per item
vis_df1 = vis_df[vis_df['annotations_per_item'] != 10]
```

```python
# set font and theme
sns.reset_defaults()
sns.set_context("notebook", font_scale=1.4)

plt.figure(figsize=(9, 6))

# pivot to show accuracy by annotators per item and subpop size
pivot_table = vis_df1.pivot(index='annotations_per_item', columns='subpopulation_size', values='accuracy')

# create heatmap
g = sns.heatmap(pivot_table,  annot=True, cmap="flare_r", cbar_kws={'label': 'Accuracy'})

# rename axis labels
g.set_xlabel('Subpopulation Size')
g.set_ylabel('Annotators Per Item')
g.invert_yaxis()

plt.savefig("subpopulation_size.pdf", format="pdf", bbox_inches="tight")
plt.show()
```
